### ① 印出最新文章的「作者」「標題」「時間」

```python
'''
Your Code
'''
import requests
from bs4 import BeautifulSoup

url = 'https://www.ptt.cc/bbs/NBA/index.html'
r = requests.get(url)
soup = BeautifulSoup(r.text, 'html5lib')

posts = []

for d in soup.find_all(class_='title'):
    title= d.text.replace('\n', '').replace('\t', '')
    print("title:", title)
    try:
        post = {}

        soup2 = BeautifulSoup(requests.get('https://www.ptt.cc{}'.format(d.find('a')['href'])).text, 'html5lib')
        print('https://www.ptt.cc{}'.format(d.find('a')['href']))
        print(soup2.find_all(class_="article-meta-tag"), ":", soup2.find_all(class_="article-meta-value"))
        print(soup2.find_all(class_="article-meta-tag")[0].text, ":", soup2.find_all(class_="article-meta-value")[0].text)
        print(soup2.find_all(class_="article-meta-tag")[2].text, ":", soup2.find_all(class_="article-meta-value")[2].text)
        print(soup2.find_all(class_="article-meta-tag")[3].text, ":", soup2.find_all(class_="article-meta-value")[3].text)
        print("")
        
        post['標題'] = soup2.find_all(class_="article-meta-value")[0].text
        post['作者'] = soup2.find_all(class_="article-meta-value")[2].text
        post['時間'] = soup2.find_all(class_="article-meta-value")[3].text
        
        posts.append(post)
        
        
    except:
        continue
posts = sorted(posts, key= lambda x: x['時間'])
print(posts[-1])
```

### ② 印出第一頁所有文章的「作者」「標題」「時間」

```python
'''
Your Code
'''

url = "https://www.ptt.cc/bbs/Stock/index.html"
r = requests.get(url)
soup = BeautifulSoup(r.text, 'html5lib')
for d in soup.find_all(class_='r-ent'):
    print("作者:", d.find(class_='author').text.replace('\n','').replace('\t', ''))
    print("標題:", d.find(class_='title').text.replace('\n','').replace('\t', ''))
    print("時間:", d.find(class_='date').text.replace('\n','').replace('\t', ''))
    print("")
```

### ③ 試著爬爬看其他版的文章

```python
'''
Your Code
'''

url = "https://www.ptt.cc/bbs/Beauty/index.html"
r = requests.get(url, cookies={'over18': '1'})
soup = BeautifulSoup(r.text, 'html5lib')

for d in soup.find_all(class_='r-ent'):
    try:
        print("作者:", d.find(class_='author').text.replace('\n','').replace('\t', ''))
        print("標題:", d.find(class_='title').text.replace('\n','').replace('\t', ''))
        print('路徑:', "https://www.ptt.cc{}".format(d.find('a')['href']))
        print("時間:", d.find(class_='date').text.replace('\n','').replace('\t', ''))
        print("")
    except:
        continue
```
