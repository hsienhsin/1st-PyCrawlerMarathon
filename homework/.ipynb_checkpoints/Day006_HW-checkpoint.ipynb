## 作業目標

* 根據範例提供的 API ，完成以下問題：

    1. 取出知乎問題發問時間
    2. 取出第一筆與最後一筆回答的時間


```python
import requests
headers = {'user-agent': 'my-app/0.0.1'}
r = requests.get('https://www.zhihu.com/api/v4/questions/55493026/answers', headers=headers)
response = r.text
print(type(response))
print(response[:100])
```

```python
import json
data = json.loads(response)

print(json.dumps(data, indent=3))

for d in data['data']:
    print(d)
```

### 1. 取出知乎問題發問時間

```python
from datetime import datetime

#utc_time = datetime.fromtimestamp(1486390229, timezone.utc)
#local_time = utc_time.astimezone()
#print(local_time.strftime("%Y-%m-%d %H:%M:%S.%f%z (%Z)"))
```

```python
# 1. 取出知乎問題發問時間

'''
Your Code
'''
question_time = data['data'][0]['question']['created']
print(question_time)

question_time_date = datetime.fromtimestamp(question_time)
print(question_time_date)
```

### 2. 取出第一筆與最後一筆回答的時間

```python
# 2. 取出第一筆與最後一筆回答的時間

'''
Your Code
'''
timeList = []
for d in data['data']:
    timeList.append(d['created_time'])
print(timeList)
timeList.sort()
print(timeList)


local_time = datetime.fromtimestamp(timeList[0])
print(timeList[0], ", ", local_time)

local_time = datetime.fromtimestamp(timeList[-1])
print(timeList[-1], ", ", local_time)
```
